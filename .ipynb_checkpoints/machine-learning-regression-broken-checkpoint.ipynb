### Exploring the Factors Impacting a Movie's Profitability

```python
import pandas as pd
from sklearn.model_selection import GridSearchCV, train_test_split

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder

from sklearn.compose import ColumnTransformer

from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import SGDRegressor

from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import ExtraTreeRegressor

from sklearn.svm import SVR
from sklearn.svm import LinearSVR

from sklearn.neural_network import MLPRegressor

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

from sklearn.pipeline import Pipeline

import matplotlib.pyplot as plt
%matplotlib inline
```

The prepared data is imported:

```python
save_dir = "../../Output/"
imdb_dummies_df = pd.read_csv(save_dir + "imdb_dummies_df_2020_09_15.csv", sep="|")
print("imdb_dummies_df.shape :", imdb_dummies_df.shape)
cols = list(imdb_dummies_df.columns)
print("cols :", cols[:10])
```

The target variable is created (y).

```python
ml_df = imdb_dummies_df.copy()
ml_df["profitability"] = (ml_df["revenue"] -  ml_df["budget"]) / ml_df["budget"]
print("ml_df.shape :", ml_df.shape)
ml_df.head()
```

```python
budget_threshold = 1000000
profitability_threshold = 500000
ml_df = ml_df[(ml_df["budget"] > budget_threshold) & \
              (ml_df["profitability"] < budget_threshold)]
ml_df.shape
```

The variabels called X and y are prepared for machine learning.

```python
ml_df["startYear"] = ml_df["startYear"].astype(int).astype(str)
drop_cols = ["primaryTitle", "originalTitle", "imdb_id", "revenue", "budget"]
ml_df.drop(drop_cols, axis="columns", inplace=True)
ml_df.set_index("tconst", inplace=True)
# print(ml_df[list(ml_df.columns)[:5]].head())

y = ml_df["profitability"]
X = ml_df.drop(["profitability"], axis="columns")

features = list(X.columns)
print("len(features) :", len(features))
if "revenue" in features:
    print("revenue in features!")

# print("X.dtypes :", X.dtypes)
X.head()
```

To check if the transformations worked as intended, we check the shape of the X and y variables.

```python
print("y.shape :", y.shape)
print("X.shape :", X.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print("X_train.shape : {}, X_test.shape : {}, y_train.shape : {}, y_test.shape : {}" \
      .format(X_train.shape, X_test.shape, y_train.shape, y_test.shape))
```

```python
rnd_st = 54125
```

```python
# Linear Regression
t = [("num", MinMaxScaler(), ["runtime"])]#, ("cat", OneHotEncoder(), ["startYear"])]
col_transform = ColumnTransformer(transformers=t)

pipeline = Pipeline(steps=[('prep', col_transform), ('estimator', SVR())])

grid_dict = [{"estimator": [Lasso(max_iter=10000)],
              "estimator__alpha": [0.1, 1.0, 10.0, 100.0], 
              "estimator__max_iter": [10000],
              "estimator__random_state": [rnd_st]
             },
#              {"estimator": [Ridge()],
#               "estimator__alpha": [0.1, 1.0, 10.0, 100.0], 
#               "estimator__max_iter": [10000],
#               "estimator__solver": ["svd", "cholesky", "lsqr", "sparse_cg", "sag", "saga"],
#               "estimator__random_state": [rnd_st]
#              },
#              {"estimator": [ElasticNet()],
#               "alpha": [0.1, 1.0, 10.0, 100.0], 
#               "esti"
#               "max_iter": [10000],
#               "random_state": [rnd_st]
#              },
#              {"estimator": [SGDRegressor()],
#               "alpha": [0.1, 1.0, 10.0, 100.0], 
#               "max_iter": [10000],
#               "random_state": [rnd_st]
#              }
            ]

grid = GridSearchCV(estimator=pipeline, param_grid=grid_dict, scoring="r2", cv=5) # , n_jobs=2

grid.fit(X_train, y_train)
best_params = grid.best_params_
print("best_params :", best_params)

y_pred = grid.predict(X_test)
score = grid.score(X_test, y_test)
print("score :", score)
mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
print("mae :", mae)

plt.clf()
plt.scatter(y_pred, y_test, marker="o", color="blue")
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
# plt.savefig(save_dir + "Predicted_vs_Actual.png")
plt.show()
```

```python
# Support Vector Machine
est = SVR()

print("est.get_params().keys() ", est.get_params().keys())
grid_dict = {"kernel": ["linear"],  # "linear", ‘poly’, ‘rbf’, ‘sigmoid’"
             "degree": [1, 2, 3],
             "C": [0.1, 1.0, 10.0],
             "gamma": ["scale", "auto"]
             }
# grid_dict = {}
grid = GridSearchCV(estimator=est,
                    param_grid=grid_dict,
                    scoring="r2",
                    n_jobs=6,
                    cv=5)

grid.fit(X_train, y_train)
best_params = grid.best_params_
print("best_params :", best_params)

y_pred = grid.predict(X_test)
score = grid.score(X_test, y_test)
print("score :", score)
mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
print("mae :", mae)

plt.clf()
plt.scatter(y_pred, y_test, marker="o", color="blue")
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
# plt.savefig(save_dir + "Predicted_vs_Actual.png")
plt.show()
```
