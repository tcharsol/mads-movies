## Data Exploration

```python
import pandas as pd
import numpy as np
import eda 
import matplotlib.pyplot as plt
import re
```

```python
movielens = pd.read_csv('results.csv')
```

```python
#movielens['revenue'] = pd.to_numeric(movielens['revenue'])
#movielens['budget'] = movielens['budget'].astype('float64')
#movielens['profit_ratio'] = movielens['revenue']/movielens['budget']

movielens['revenue_isnumeric'] = movielens['revenue'].str.isnumeric()
movielens['budget_isnumeric'] = movielens['budget'].str.isnumeric()
movielens_con = movielens[(movielens['revenue_isnumeric'] == True) & (movielens['budget_isnumeric'] == True)]

movielens_con['budget'] = pd.to_numeric(movielens_con['budget'])
movielens_con['revenue'] = pd.to_numeric(movielens_con['revenue'])
movielens_con = movielens_con[movielens_con['revenue'] > 0]
movielens_con = movielens_con[movielens_con['budget'] > 10000]
movielens_con['profit_ratio'] = movielens_con['revenue']/movielens_con['budget']
movielens_con[['revenue','budget','profit_ratio']]
```

```python
#movielens_con.dropna(subset=['profit_ratio'],inplace=True)
movielens_con
```

```python
eda.run_preliminary_EDA(movielens_con)
```

```python
movielens_con.sort_values(by=['profit_ratio'],ascending=False)[['genres','title','revenue','budget','profit_ratio']]
```
